```python
def get_year_quarterly_data_2018( dataset , typeofstatement , date , companysymbol): 
    companyname_data = quandl.get_table(dataset , dimension = typeofstatement , calendardate = {'gte': date}, ticker = companysymbol)
    companyname_data_filtered = companyname_data.filter(items = ['calendardate', 'currentratio', 'assets', 'assetsavg', 'assetsc', 'cashenq', 'liabilitiesc', 'debt' , 'debtc' , 'debtnc', 'revenue' , 'capex','intexp', 'ebit', 'optinc', 'netinc','net', 'equityusd', 'roa', 'roe', 'de', 'gp', 'eps'])
    companyname_data_year1 = companyname_data_filtered.iloc[7:]
    return companyname_data_year1 

def get_year_quarterly_data_2019( dataset , typeofstatement , date , companysymbol): 
    companyname_data = quandl.get_table(dataset , dimension = typeofstatement , calendardate = {'gte': date}, ticker = companysymbol)
    companyname_data_filtered = companyname_data.filter(items = ['calendardate', 'currentratio', 'assets', 'assetsavg', 'assetsc', 'cashenq', 'liabilitiesc', 'debt' , 'debtc' , 'debtnc', 'revenue' , 'capex','intexp', 'ebit', 'optinc', 'netinc','net', 'equityusd', 'roa', 'roe', 'de', 'gp', 'eps'])
    companyname_data_year2 = companyname_data_filtered.iloc[3:7]
    return companyname_data_year2

def get_year_quarterly_data_2020(  dataset , typeofstatement , date , companysymbol): 
    companyname_data = quandl.get_table(dataset , dimension = typeofstatement , calendardate = {'gte': date}, ticker = companysymbol)
    companyname_data_filtered = companyname_data.filter(items = ['calendardate', 'currentratio', 'assets', 'assetsavg', 'assetsc', 'cashenq', 'liabilitiesc', 'debt' , 'debtc' , 'debtnc', 'revenue' , 'capex','intexp', 'ebit', 'optinc', 'netinc','net', 'equityusd', 'roa', 'roe', 'de', 'gp', 'eps'])
    companyname_data_year3 = companyname_data_filtered.iloc[0:3]
    return companyname_data_year3


def current_ratio(companies = []): 
    a = ['ca1' , 'ca2' , 'ca3', 'ca4', 'ca5' , 'ca6']
    b = ['ca11' , 'ca22', 'ca33', 'ca44', 'ca55', 'ca66']
    c = ['ca111' , 'ca222', 'ca333', 'ca444', 'ca555', 'ca666']
    list_1 = []
    list_2 = []
    list_3 = []
    for ratio in a: 
        for company in companies: 
            ratio = company 
            list_1.append(ratio)
    for ratio_2 in b:
        for new in list_1: 
            ratio_2 = new.filter(items = ['calendardate', 'currentratio'])
            list_2.append(ratio_2)
    for ratio_3 in c: 
        for z in list_2: 
            ratio_3 = z.set_index('calendardate')
            list_3.append(ratio_3)
    combined = pd.concat([x for x in list_3], axis = 1 , ignore_index = True)
    combined2 = combined.filter(items=[0 ,1,2,3,4,5])
    return combined2


def debt_to_equity(companies = []): 
    a = ['de1' , 'de2' , 'de3', 'de4', 'de5' , 'de6']
    b = ['de11' , 'de22', 'de33', 'de44', 'de55', 'de66']
    c = ['de111' , 'de222', 'de333', 'de444', 'de555', 'de666']
    list_1 = []
    list_2 = []
    list_3 = []
    for ratio in a: 
        for company in companies: 
            ratio = company 
            list_1.append(ratio)
    for ratio_2 in b:
        for new in list_1: 
            ratio_2 = new.filter(items = ['calendardate', 'de'])
            list_2.append(ratio_2)
    for ratio_3 in c: 
        for z in list_2: 
            ratio_3 = z.set_index('calendardate')
            list_3.append(ratio_3)
    combined = pd.concat([x for x in list_3], axis = 1 , ignore_index = True)
    combined2 = combined.filter(items=[0 ,1,2,3,4,5])
    return combined2
    


def debt_to_assets(companies = []): 
    a = ['de1' , 'de2' , 'de3', 'de4', 'de5' , 'de6']
    b = ['de11' , 'de22', 'de33', 'de44', 'de55', 'de66']
    c = ['de111' , 'de222', 'de333', 'de444', 'de555', 'de666']
    d = ['de1111' , 'de2222', 'de3333', 'de4444', 'de5555', 'de6666']
    list_1 = []
    list_2 = []
    list_3 = []
    list_4 = []
    for ratio in a: 
        for company in companies: 
            ratio = company
            list_1.append(ratio)
    for ratio_2 in b: 
        for new in list_1: 
            ratio_2 = new.set_index('calendardate')
            list_2.append(ratio_2)
    for ratio_3 in c: 
        for z in list_2: 
            z['D/A'] = (z["debt"]/z['assets'])
            ratio_3 = z.filter(items = ['D/A'])
            list_3.append(ratio_3)
    combined1 = pd.concat([l for l in list_3], axis = 1 , ignore_index = True)
    combined2 = combined1.filter(items = [0,1,2,3,4,5])
    return combined2
                        

def interest_coverage(companies = []): 
    a = ['de1' , 'de2' , 'de3', 'de4', 'de5' , 'de6']
    b = ['de11' , 'de22', 'de33', 'de44', 'de55', 'de66']
    c = ['de111' , 'de222', 'de333', 'de444', 'de555', 'de666']
    d = ['de1111' , 'de2222', 'de3333', 'de4444', 'de5555', 'de6666']
    list_1 = []
    list_2 = []
    list_3 = []
    list_4 = []
    for ratio in a: 
        for company in companies: 
            ratio = company
            list_1.append(ratio)
    for ratio_2 in b: 
        for new in list_1: 
            ratio_2 = new.set_index('calendardate')
            list_2.append(ratio_2)
    for ratio_3 in c: 
        for z in list_2: 
            z['Interest Coverage'] = (z["ebit"]/z['intexp'])
            ratio_3 = z.filter(items = ['Interest Coverage'])
            list_3.append(ratio_3)
    combined1 = pd.concat([l for l in list_3], axis = 1 , ignore_index = True)
    combined2 = combined1.filter(items = [0,1,2,3,4,5])
    return combined2

```
