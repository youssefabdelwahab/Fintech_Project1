```python
import pandas as pd
import datetime as dt
```

```python
USmobility_data = pd.read_csv('2020_US_Region_Mobility_Report.csv',index_col ='date',parse_dates = True,infer_datetime_format = True)
```

```python
USmobility_data.head()
```

```python
# replacing all empty sub-regions which represent the entire US with code 'ALL'
USmobility_data['sub_region_1'].fillna(value ='ALL', inplace=True)
```

```python
# checking that it worked
USmobility_data
```

```python
USmobility_data.columns
```

```python
USmobility_data.drop(columns = ['metro_area', 'census_fips_code'])
```

```python
# grouping the data into Q's for all of the US
# 1) starting out with the entire timeframe for retail and recreation column 
US_retail=USmobility_data.loc[USmobility_data.sub_region_1 == 'ALL','retail_and_recreation_percent_change_from_baseline']
```

```python
US_retail.plot()
```

```python
# using count to understand how many rows US-ALL includes to be able to divide into Q's
# this is the quartly sum of percent changes 
US_Quaters = US_retail.groupby(pd.Grouper(freq="3M")).mean()
US_Quaters.head()
```

```python
US_Quaters.plot()
```

```python
#plotting based on sum of % change 
US_Quaters = US_retail.groupby(pd.Grouper(freq="3M")).sum().plot()
```
